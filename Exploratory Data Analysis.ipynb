## Write a function to display an image and the bounding boxes

Implement the `display_images` function below. This function takes a batch as an input and display an image with its corresponding bounding boxes. The only requirement is that the classes should be color coded (eg, vehicles in red, pedestrians in blue, cyclist in green).

```python

def display_images(batch):
    #colorcode the figures
    colormap = {1:'red', 2:'blue', 4:'green'} #1 = vehicles, 2 = pedestrians, 4 = cyclist

    for tf_data in batch:
        fig, ax = plt.subplots()
        img = tf_data["image"].numpy()
        ax.imshow(img)
        classes = tf_data["groundtruth_classes"].numpy()
        box_coords =tf_data["groundtruth_boxes"].numpy()
        w = img.shape[1]
        h = img.shape[0]

        for iterator, coords in enumerate(box_coords):
            cur_class = classes[iterator]
            xy = (w*coords[1], h*coords[0]) #tuple for the corner xy coord of the box
            width = w * (coords[3]-coords[1])
            height =h * (coords[2]-coords[0])

            rect = patches.Rectangle(xy=xy, width=width, height=height ,edgecolor=colormap[cur_class],facecolor='none')
            ax.add_patch(rect)

        plt.show()


```

## Display 10 images 

Using the dataset created in the second cell and the function you just coded, display 10 random images with the associated bounding boxes. You can use the methods `take` and `shuffle` on the dataset.

```python
# Display 10 random images in dataset

num_to_display= 10
sample_batch = dataset.shuffle(42, reshuffle_each_iteration=True).take(num_to_display)

display_images(sample_batch)
```

## Additional EDA

In this last part, you are free to perform any additional analysis of the dataset. What else would like to know about the data?
For example, think about data distribution. So far, you have only looked at a single file...

```python
## Taking a look at the Class distribution
import seaborn as sns

print('Class distribution in the dataset')
count_each_class = {1: 0, 2:0, 4: 0}
cars = 0
people = 0
cyclist = 0
count_class_per_img  = {1:[], 2:[], 4:[]}
all_imgs = 0

for img in dataset.take(50000):
    all_imgs += 1
    classes = img['groundtruth_classes'].numpy()
    values, counts = np.unique(classes, return_counts=True)
    #update the count of each class after getting the full count per image
    for i in range(len(values)):
        count_each_class[values[i]] = count_each_class[values[i]] + counts[i]
        count_class_per_img[values[i]].append(counts[i])
```

```python
print("Distribution analysis for 50000 frames")

x = ["cars", "pedestrians", "cyclists"]
y = list(count_each_class.values())

ax = sns.barplot(x=x, y=y)
ax.set(xlabel='Classes', ylabel='Number of total objects', title='Distrubution of the classes over 50k datapoints')

print("Distribution analysis of individual class representation per image")


plt.figure(figsize=(10, 10))
ax_car = sns.histplot(data=count_class_per_img[1], kde=True)
ax_car.set(xlabel='Number of cars per image', ylabel='Number of images', title='A distribution of the number of cars in datapoints over in 50k datapoints')


plt.figure(figsize=(10, 10))
ax_car = sns.histplot(data=count_class_per_img[2], kde=True)
ax_car.set(xlabel='Number of pedestrians per image', ylabel='Number of images', title='A distribution of the number of pedestrians over 50k datapoints')
plt.show()

plt.figure(figsize=(10, 10))
ax_bicycle = sns.histplot(data=count_class_per_img[4])
ax_bicycle.set(xlabel='Number of cyclists per image', ylabel='Number of images', title='A distribution of the number of cyclists over in 50k datapoints')
plt.show()
```
